#MARCO TEÓRICO

Un modelo seq2seq (del inglés "sequence-to-sequence") es un tipo de modelo de aprendizaje automático que se utiliza a menudo en el procesamiento del lenguaje natural. Estos modelos consisten en dos partes principales: un codificador que convierte la entrada en una representación interna, y un decodificador que utiliza esta representación interna para producir una salida. Esta estructura es útil para tareas como la traducción de texto, ya que permite que el modelo aprenda a asociar una secuencia de entrada (como una frase en un idioma) con una secuencia de salida (como una traducción a otro idioma). Los modelos seq2seq pueden ser implementados con diferentes tipos de redes neuronales, como redes recurrentes o modelos transformers [1].
Los modelos transformers se han vuelto muy populares en los últimos años en el campo del procesamiento del lenguaje natural, incluyendo en la traducción de texto. Los modelos transformers utilizan una arquitectura de red neuronal que permite que el modelo considere la totalidad de la entrada de una sola vez al hacer predicciones, en lugar de procesarla en fragmentos secuenciales como lo hacen muchos otros modelos. Esto puede resultar en una traducción de texto más precisa y eficiente. Además, los modelos transformers son entrenables en una sola etapa, lo que los hace más rápidos y fáciles de entrenar que otros modelos de traducción de texto [2].

```python
class AttentionLayer(tf.keras.layers.Layer):
    """
     Esta clase implementa la atención de Bahdanau (https://arxiv.org/pdf/1409.0473.pdf).
     Hay tres conjuntos de pesos introducidos W_a, U_a y V_a
     """

    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        assert isinstance(input_shape, list)
        # Create a trainable weight variable for this layer.
        """
1. Usamos la función add_weight para crear las matrices de peso entrenables que necesitamos.
El primer argumento es el nombre de la matriz de peso, el segundo es la forma de la matriz, 
l tercer argumento es una cadena que especifica el inicializador de la matriz de peso y el último
 argumento es un valor booleano que especifica si la matriz de peso es entrenable o no.

2. La forma de W_a y U_a es (input_shape[0][2], input_shape[0][2]). 
La forma de entrada es una tupla que contiene las formas de las entradas. 
El primer elemento de la tupla contiene la forma de la primera entrada 
(en este caso, la palabra incrustaciones) y el segundo elemento de la tupla 
contiene la forma de la segunda entrada (en este caso, el vector de contexto). 
input_shape[0][2] es el número de características en la primera entrada 
(el número de características en las incrustaciones de palabras). 
input_shape[1][2] es el número de características en la segunda entrada 
(el número de características en el vector de contexto). En este caso, el 
número de características en las incrustaciones de palabras y el vector de 
contexto es el mismo. [3]
"""
        self.W_a = self.add_weight(name='W_a',
                                   shape=tf.TensorShape((input_shape[0][2], input_shape[0][2])),
                                   initializer='uniform',
                                   trainable=True)
        self.U_a = self.add_weight(name='U_a',
                                   shape=tf.TensorShape((input_shape[1][2], input_shape[0][2])),
                                   initializer='uniform',
                                   trainable=True)
        self.V_a = self.add_weight(name='V_a',
                                   shape=tf.TensorShape((input_shape[0][2], 1)),
                                   initializer='uniform',
                                   trainable=True)

        super(AttentionLayer, self).build(input_shape)  # Be sure to call this at the end

    def call(self, inputs):
        """
        inputs: [encoder_output_sequence, decoder_output_sequence]
        """
        assert type(inputs) == list
        encoder_out_seq, decoder_out_seq = inputs

        logger.debug(f"encoder_out_seq.shape = {encoder_out_seq.shape}")
        logger.debug(f"decoder_out_seq.shape = {decoder_out_seq.shape}")

        def energy_step(inputs, states):
            """ Step function for computing energy for a single decoder state
            inputs: (batchsize * 1 * de_in_dim)
            states: (batchsize * 1 * de_latent_dim)
            """

            logger.debug("Running energy computation step")

            if not isinstance(states, (list, tuple)):
                raise TypeError(f"States must be an iterable. Got {states} of type {type(states)}")

            encoder_full_seq = states[-1]

            """ Computing S.Wa where S=[s0, s1, ..., si]"""
            # <= batch size * en_seq_len * latent_dim
            W_a_dot_s = K.dot(encoder_full_seq, self.W_a)

            """ Computing hj.Ua """
            U_a_dot_h = K.expand_dims(K.dot(inputs, self.U_a), 1)  # <= batch_size, 1, latent_dim

            logger.debug(f"U_a_dot_h.shape = {U_a_dot_h.shape}")

            """ tanh(S.Wa + hj.Ua) """
            # <= batch_size*en_seq_len, latent_dim
            Ws_plus_Uh = K.tanh(W_a_dot_s + U_a_dot_h)

            logger.debug(f"Ws_plus_Uh.shape = {Ws_plus_Uh.shape}")

            """ softmax(va.tanh(S.Wa + hj.Ua)) """
            # <= batch_size, en_seq_len
            e_i = K.squeeze(K.dot(Ws_plus_Uh, self.V_a), axis=-1)
            # <= batch_size, en_seq_len
            e_i = K.softmax(e_i)

            logger.debug(f"ei.shape = {e_i.shape}")

            return e_i, [e_i]

        def context_step(inputs, states):
            """ Step function for computing ci using ei """

            logger.debug("Running attention vector computation step")

            if not isinstance(states, (list, tuple)):
                raise TypeError(f"States must be an iterable. Got {states} of type {type(states)}")

            encoder_full_seq = states[-1]

            # <= batch_size, hidden_size
            c_i = K.sum(encoder_full_seq * K.expand_dims(inputs, -1), axis=1)

            logger.debug(f"ci.shape = {c_i.shape}")

            return c_i, [c_i]

        # we don't maintain states between steps when computing attention
        # attention is stateless, so we're passing a fake state for RNN step function
        fake_state_c = K.sum(encoder_out_seq, axis=1)
        fake_state_e = K.sum(encoder_out_seq, axis=2)  # <= (batch_size, enc_seq_len, latent_dim

        """ Computing energy outputs """
        # e_outputs => (batch_size, de_seq_len, en_seq_len)
        last_out, e_outputs, _ = K.rnn(
            energy_step, decoder_out_seq, [fake_state_e], constants=[encoder_out_seq]
        )

        """ Computing context vectors """
        last_out, c_outputs, _ = K.rnn(
            context_step, e_outputs, [fake_state_c], constants=[encoder_out_seq]
        )

        return c_outputs, e_outputs

    def compute_output_shape(self, input_shape):
        """ Outputs produced by the layer """
        return [
            tf.TensorShape((input_shape[1][0], input_shape[1][1], input_shape[1][2])),
            tf.TensorShape((input_shape[1][0], input_shape[1][1], input_shape[0][1]))
        ]
```

```python
"""
1. Usaremos un LSTM simple con 300 unidades. Usaremos 2 capas de LSTM. Cada LSTM devolverá secuencias y estados.
2. El codificador será un LSTM de 3 capas. Las primeras 2 capas devolverán secuencias y estados, mientras 
que la última capa devolverá tanto secuencias como estados.
3. El decodificador será un LSTM de 1 capa. Devolverá secuencias y estados.
4. La capa de Atención tomará las salidas del codificador y las salidas del decodificador como entradas. 
Devolverá tanto salidas como estados.
5. Concatenaremos las salidas del decodificador y las salidas de atención usando una capa Concatenar.
6. La salida del decodificador se alimentará a la capa TimeDistributed Dense con activación softmax. """
K.clear_session()

latent_dim = 300 
embedding_dim = 100 

encoder_inputs = Input(shape=(max_eng_len,))
encoder_emb = Embedding(eng_vocab_size, embedding_dim, trainable=True)(encoder_inputs)

encoder_lstm1 = LSTM(latent_dim, return_sequences=True, return_state=True, dropout=0.3, recurrent_dropout=0.2)
encoder_output1, state_h1, state_c1 = encoder_lstm1(encoder_emb)

encoder_lstm2 = LSTM(latent_dim, return_sequences=True, return_state=True, dropout=0.3, recurrent_dropout=0.2)
encoder_output2, state_h2, state_c2 = encoder_lstm2(encoder_output1)

encoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True, dropout=0.3, recurrent_dropout=0.2)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_output2)
encoder_states = [state_h,state_c]

decoder_inputs = Input(shape=(None,))
decoder_emb = Embedding(ger_vocab_size, embedding_dim, trainable=True)(decoder_inputs)

decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True, dropout=0.3, recurrent_dropout=0.2)
decoder_outputs, decoder_fwd_state, decoder_bwd_state = decoder_lstm(decoder_emb, initial_state=encoder_states) 

attn_layer = AttentionLayer(name='attention_layer') 
attn_outputs, attn_states = attn_layer([encoder_outputs, decoder_outputs]) 

decoder_concat_outputs = Concatenate(axis=-1,name='concat_layer')([decoder_outputs, attn_outputs])

decoder_dense = TimeDistributed(Dense(ger_vocab_size,activation='softmax'))
decoder_outputs = decoder_dense(decoder_concat_outputs)

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='rmsprop',metrics='accuracy')
"""
1. Usamos la función de pérdida de entropía cruzada para calcular la pérdida entre el valor predicho y el valor real.
2. Usamos el optimizador RMSProp para optimizar los pesos.
3. usamos la precisión como la métrica. """
```

```python
 X_train, X_test, y_train, y_test = train_test_split(eng_pad_sequences,ger_pad_sequences,test_size=0.25,shuffle=True,random_state=101)
 """
 la función train_test_split del módulo sklearn.model_selection para dividir un
  conjunto de datos en dos subconjuntos: uno para entrenamiento y otro para pruebas. 
  Los datos consisten en secuencias rellenadas de oraciones en inglés y alemán, y 
  se está pasando la secuencia de oraciones alemanas como la secuencia de destino para el modelo. 
  Se está estableciendo el tamaño del conjunto de pruebas en 25% y se está mezclando los 
  datos aleatoriamente antes de dividirlos. Además, se está fijando la semilla del generador
  de números aleatorios para obtener resultados consistentes en cada ejecución.
 """
```

```python
es = EarlyStopping(monitor='val_accuracy',mode='max',verbose=1,patience=40)
rl = ReduceLROnPlateau(monitor='val_accuracy',mode='max',verbose=1,patience=5,factor=0.1,min_lr=0.001)
mc = ModelCheckpoint('checkpoint/',monitor='val_accuracy',verbose=1,mode='max',save_best_only=True)
"""
1. EarlyStopping: Early Stopping se utiliza para detener el proceso de entrenamiento si el modelo deja de aprender.
     Aquí, monitoreamos val_accuracy (precisión de validación).
     Si no mejora la precisión durante 40 épocas, se detendrá el proceso de entrenamiento.
2. ReduceLROnPlateau: esta devolución de llamada reduce la tasa de aprendizaje en un factor de 0,1 si la precisión de la validación no mejora durante 5 épocas.
3. ModelCheckpoint: esta devolución de llamada guarda el modelo con la mejor precisión de validación.
4. r: Esta es la función model.fit(). Se utiliza para entrenar el modelo.
     Usamos 5 épocas, un tamaño de lote de 256, y usamos las devoluciones de llamada que definimos anteriormente.
"""
r = model.fit([X_train,y_train[:,:-1]],
             y_train.reshape(y_train.shape[0],y_train.shape[1],1)[:,1:],
             epochs=5,
             batch_size=256,
             callbacks=[es,rl,mc],
             validation_data=([X_test,y_test[:,:-1]],y_test.reshape(y_test.shape[0],y_test.shape[1],1)[:,1:]))
```

```python
model.save('english_to_german_translator.h5')
lm = load_model('english_to_german_translator.h5',custom_objects={'AttentionLayer': AttentionLayer})
"""
cargamos el modelo, se cargan los pesos del modelo en una nueva variable llamada lm. 
Luego, se crea una instancia de la clase AttentionLayer(), que se reafirma como el 
mecanismo de atención del modelo. Finalmente, se carga el modelo necesitará la 
instancia de la clase AttentionLayer() como el parámetro custom_objects.

Nota al pie: Necesitamos usar el mismo objeto de clase que se usó para el entrenamiento del modelo. 
Esto se debe a que Keras guarda el nombre de la clase y luego lo carga usando el parámetro custom_objects 
al cargar el modelo. """
```

```python
"""
obtenemos las salidas del codificador y los estados del modelo del codificador. 
Estos estados se utilizan como estados iniciales para el modelo del decodificador. 
Las salidas del decodificador se procesan en la capa de atención, que devuelve un vector 
de contexto que se concatena con las salidas del decodificador. Finalmente, la capa densa 
final procesa esta información y devuelve la predicción, que se utiliza como entrada para el 
modelo del decodificador en el siguiente paso de tiempo.
"""

encoder_model = Model(inputs=encoder_inputs, outputs=[encoder_outputs, state_h, state_c])

decoder_initial_state_a = Input(shape=(latent_dim,))
decoder_initial_state_c = Input(shape=(latent_dim,))
decoder_hidden_state = Input(shape=(max_eng_len, latent_dim))

decoder_out, decoder_state_a, decoder_state_c = decoder_lstm(decoder_emb, initial_state=[decoder_initial_state_a, decoder_initial_state_c])
attn_out_inf, attn_states_inf = attn_layer([decoder_hidden_state, decoder_out])
decoder_inf_concat_outputs = Concatenate(axis=-1, name='concat')([decoder_out, attn_out_inf])

decoder_final = decoder_dense(decoder_inf_concat_outputs)
decoder_model = Model([decoder_inputs]+[decoder_hidden_state, decoder_initial_state_a, decoder_initial_state_c], [decoder_final]+[decoder_state_a, decoder_state_c])
```

```python

def decode_sequences(input_sequence):
    encoder_out, encoder_a, encoder_c = encoder_model.predict(input_sequence)
    next_input = np.zeros((1,1))
    next_input[0,0] = ger_tokenizer.word_index['start']
    output_sequence = ''
    flag = False
    while not flag:
        decoded_out, trans_state_a, trans_state_c = decoder_model.predict([next_input] + [encoder_out, encoder_a, encoder_c])
        output_idx = np.argmax(decoded_out[0, -1, :])
        if output_idx == ger_tokenizer.word_index['end']: 
            flag = True
        elif output_idx > 0 and output_idx != ger_tokenizer.word_index['start']:
            output_token = ger_tokenizer.index_word[output_idx] 
            output_sequence = output_sequence + ' ' + output_token 
        next_input[0,0] = output_idx
        # Continously update the transient state vectors in decoder.
        encoder_a, encoder_c = trans_state_a, trans_state_c
        
    return output_sequence     
    """
Se han creado los modelos de codificador y decodificador en una sección anterior.
 Ahora, se están utilizando estos modelos para predecir la salida del codificador, 
los estados del codificador y los estados del decodificador. Luego, se crea una matriz 
numérica de ceros como la siguiente entrada para el decodificador y se configura el valor 
de esta entrada para el token de inicio. Se inicializa una secuencia de salida vacía y se 
establece una bandera en False. Se utiliza un bucle while para predecir la secuencia de
 salida, y se comprueba si el índice del token de máxima probabilidad es el token final.
 Si lo es, se establece la bandera en True y se sale del bucle. Si el índice del token 
de máxima probabilidad es válido, se agrega el token correspondiente a la secuencia de
 salida y se actualizan los estados del decodificador. Finalmente, se devuelve la 
secuencia de salida.
"""
```

```python
def sequence_to_language(input_sequence,tokenizer,lang):
    res = ''
    
    if lang == 'eng':
        for idx in input_sequence:
            if idx != 0:
                res = res + tokenizer.index_word[idx] + ' '
    
    elif lang == 'ger':
        for idx in input_sequence:
            if idx > 0:
                if tokenizer.index_word[idx] != 'start' and tokenizer.index_word[idx] != 'end':
                    res = res + tokenizer.index_word[idx] + ' '
                    
    return res
```

```python
def predicted_sequence_to_ger_language(input_sequence):
    res = ''
    
    for idx in input_sequence:
        if idx > 0:
            if ger_tokenizer.index_word[idx] != 'start' and ger_tokenizer.index_word[idx] != 'end':
                res = res + ger_tokenizer.index_word[idx] + ' '
    
    return res
  
"""
El código anterior describe una función que toma una secuencia de entrada, un tokenizador y un idioma.
Inicializa una cadena vacía para almacenar la salida. Si el idioma es inglés, inicia un bucle for
 para iterar sobre la secuencia de entrada y verifica si el índice actual no es 0, ya que 0 es el
 relleno para las oraciones más cortas. Si el índice no es 0, agrega la palabra correspondiente 
a ese índice a la cadena de salida y agrega un espacio. Si el idioma es alemán, inicia un bucle for 
para iterar sobre la secuencia de entrada y verifica si el índice es mayor que 0 y la palabra
 correspondiente al índice no es 'start' o 'end'. Si se cumplen estas condiciones, agrega
 la palabra a la cadena de salida y agrega un espacio. Finalmente, la función devuelve la
 cadena de salida como resultado."""  
```

#DISCUSIÓN

Un modelo seq2seq con atención es una variante de la arquitectura seq2seq que se utiliza para resolver problemas de secuencia a secuencia, como la traducción de idiomas. La arquitectura seq2seq con atención agrega un mecanismo de atención al modelo seq2seq tradicional, que permite al decodificador "prestar atención" a diferentes partes de la entrada mientras genera la salida.

Uno de los usos más comunes de un modelo seq2seq con atención es la traducción de idiomas, como del inglés al alemán. Para entrenar un modelo seq2seq con atención para esta tarea, se necesita un conjunto de pares de oraciones en inglés y alemán. El codificador se entrena para procesar la oración en inglés y generar una representación compacta de la misma, mientras que el decodificador con atención se entrena para generar la oración en alemán a partir de la representación del codificador y el mecanismo de atención. Una vez entrenado, el modelo seq2seq con atención puede utilizarse para traducir oraciones en inglés al alemán de manera automática.

Una de las ventajas de utilizar un modelo seq2seq con atención para la traducción de idiomas es que permite al decodificador "prestar atención" a diferentes partes de la entrada mientras genera la salida. Esto puede mejorar la precisión de la traducción ya que permite al modelo tener una comprensión más profunda de la entrada. Además, la arquitectura de codificador-decodificador con atención permite que el modelo seq2seq capture la dependencia entre las palabras en la oración de entrada y las palabras en la oración de salida, lo que resulta en traducciones más precisas.

Sin embargo, una desventaja de utilizar un modelo seq2seq con atención para la traducción de idiomas es que puede ser computacional[4].
